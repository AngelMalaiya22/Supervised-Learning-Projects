### Import Library 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import sklearn 
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
```

### Load Data

```python
data = pd.read_csv("creditcard.csv")
```

### Info of the Dataset 

```python
print(data.info)
```

### Description of Dataset 

```python
data.describe()
```

### head() will give first 5 rows and columns of the Dataset

```python
data.head()
```

### tail() will give last 5 rows and columns of the dataset

```python
data.tail()
```

### Checking the Missing value in the Dataset.

```python
''' Checking the Missing Value of the Dataset '''
data.isnull().sum()
```

### Droping the unnecessary row of the dataset

```python
data=data.drop("Time",axis=1)
```

### Firstly, we will drop the targets row from the dataset and so, now x contains only the features of the dataset. y contains the targets rows of the dataset.

```python
x = data.drop("Class", axis=1) #dropped targets from dataset so in x only features are remaining
y = data["Class"] #now y have all targets of dataset
```

```python
''' Features of the Dataset '''
x 
```

```python
''' Targets of the Dataset '''
y
```

### Spliting the dataset into Train Data & Test Data : test_size = 0.2 means that 80% data will be used for training purpose and remaining 20% data will be used for testing purpose. Here, random_state = 0 will make sure that on every split the data remains same

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

### Scaling / Standardizing / Normalizating the data

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test) 
```

### SMOTE (Synthetic Minority Over-sampling Technique) is initialized to handle class imbalance by generating synthetic samples of the minority class.

```python
smote = SMOTE(random_state=42)
```

### Now, SMOTE is applied to the training data to generate synthetic samples of the minority class. This balances the dataset and helps the model learn patterns from both classes effectively.

```python
x_train_sm, y_train_sm = smote.fit_resample(x_train, y_train)
```

### Comparing the Results in No. of values in class 0 and 1 Before SMOTE and After SMOTE 

```python
print("Before SMOTE:\n", y_train.value_counts())
print("\nAfter SMOTE:\n", y_train_sm.value_counts())
```

### Pipeline of SMOTE for implementation

```python
# Split → Scale → SMOTE → Train → Predict → Evaluate 
```

# Logistic Regression 

### Before SMOTE

### Creating the Model and Then, Traning the Model.

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

### y_pred contains the value which are predicted by the model after testing it through x_test features.

```python
y_pred = model.predict(x_test)
y_pred
```

### y_test contains the target values which were the part of dataset before splitting

```python
y_test
```

### Accuracy of Logistic regression Before SMOTE

```python
print("Accuracy of Logistic regression Before SMOTE : ",accuracy_score(y_pred, y_test))
```

### Classification Report of Logistic Regression Before SMOTE

```python
print("Classification Report of Logistic Regression Before SMOTE :\n",classification_report(y_test,y_pred))
```

### After SMOTE

### Training the Model Again with SMOTE values of the Dataset

```python
model.fit(x_train_sm, y_train_sm)
```

### y_pred contains the values which are predicted by the model after training with x_test feature value of the dataset 

```python
y_pred_sm = model.predict(x_test)
y_pred_sm
```

### y_test contains the target values which were the part of dataset before splitting

```python
y_test
```

### Accuracy Score of Logestic Regression After SMOTE

```python
print("Accuracy Score of Logestic Regression After SMOTE:",accuracy_score(y_pred_sm, y_test))
```

### Classification Report of Logestic Regression After SMOTE

```python
print("Classification Report of Logestic Regression After SMOTE :\n",classification_report(y_test, y_pred_sm))
```

### After applying SMOTE, the recall for the fraud class increased significantly, meaning the model was able to detect most fraudulent transactions. However, precision decreased, indicating an increase in false positives. This highlights the trade-off between detecting fraud and minimizing false alarms.

# Conclusion 

This project focused on detecting fraudulent credit card transactions using multiple machine learning models on a highly imbalanced dataset. Initial analysis showed that the dataset contained a very small number of fraudulent transactions compared to normal ones, which made the classification task challenging.

From the visualizations, it is clear that before applying SMOTE, the dataset was heavily imbalanced, causing models to achieve high accuracy while failing to detect many fraudulent cases. This was reflected in lower recall values for the fraud class.

After applying SMOTE, the class distribution became balanced, which significantly improved the recall of most models. The recall comparison graph clearly shows that models became much better at identifying fraudulent transactions after balancing the dataset. However, this improvement came with a trade-off in precision, as the number of false positives increased.

Among all the models, Random Forest performed the best after applying SMOTE, achieving a strong balance between precision and recall. The confusion matrix of the Random Forest model further confirms its effectiveness in correctly identifying both normal and fraudulent transactions.

Overall, this project demonstrates that:

* Class imbalance significantly affects model performance
* Accuracy alone is not a reliable metric in such cases
* SMOTE is an effective technique for improving fraud detection
* There is a trade-off between precision and recall
* Ensemble models like Random Forest are more robust for real-world problems

In real-world fraud detection systems, recall is more important than precision, as missing fraudulent transactions can lead to serious financial losses. This project successfully highlights the importance of proper data handling, model evaluation, and visualization in solving real-world machine learning problems.

```python
models = ["Logistic", "KNN", "Decision Tree", "Random Forest", "SVM"]

recall_before = [0.62, 0.79, 0.81, 0.78, 0.66]
recall_after = [0.94, 0.50, 0.81, 0.87, 0.91]

x = np.arange(len(models))

plt.figure()
plt.bar(x - 0.2, recall_before, 0.4, label='Before SMOTE')
plt.bar(x + 0.2, recall_after, 0.4, label='After SMOTE')

plt.xticks(x, models)
plt.ylabel("Recall")
plt.title("Recall Comparison Before vs After SMOTE")
plt.legend()
plt.show()
```
