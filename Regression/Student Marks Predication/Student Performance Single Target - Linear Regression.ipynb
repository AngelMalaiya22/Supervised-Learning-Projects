# Linear Regression Algorithm

### Import Libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb 
import sklearn 
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

### Load Dataset

```python
data1 = pd.read_csv("student_data.csv")
```

### Info of the Dataset

```python
data1.info()
```

### These columns were dropped because they are not relevant

```python
data1.drop(columns=['traveltime', 'schoolsup', 'famsup', 'paid', 'activities' , 'internet','famrel', 'romantic','famrel ', 'Dalc' , 'Walc', 'nursery', 'higher', 'sex','age','address','famsize','Pstatus',], errors='ignore', inplace=True)
data1
```

```python
data1.describe()
```

```python
data1.head()
```

### The school column contains categorical values (GP, MS) which cannot be directly used in machine learning models.To convert these categories into numerical format, we use get_dummies() function . 

```python
pd.get_dummies(data1["school"],dtype=int)
```

### Checking missing value 

```python
data1.isnull().sum() # checking missing values 
```

```python
# So, As we can see that there is no null value in the dataset 
```

### Here, G3 Represents the final grade of students , which is our Target Variable.

```python
y_data = data1["G3"]
```

### Removing the Target from the Dataset

```python
x_data = data1.drop("G3", axis=1)
```

### we apply one-hot encoding to the entire feature dataset. This converts all categorical variables into numerical (0/1) format.

```python
x_data = pd.get_dummies(x_data,dtype=int)
#This ensures that the dataset is fully prepared for machine learning models, which require numerical input.
```

```python
print(x_data)
```

### Now, we apply feature scaling using StandardScaler to normalize the input features.

```python
scaler = StandardScaler()
x_data_scaled = scaler.fit_transform(x_data)
```

### The dataset is split into training (80%) and testing (20%) sets. The training set is used to train the model, while the testing set is used to evaluate its performance with Predicted data of the Model.

```python
x_train , x_test , y_train , y_test = train_test_split(x_data_scaled , y_data , test_size=0.2 , random_state = 0)
```

### The Linear Regression model is trained using the training data to learn the relationship between input features and the target variable.

```python
modelv1 = LinearRegression()
modelv1.fit(x_train , y_train)
```

### Now, y_pred have the Predicted values of the model . So, now these predicted are used to evaluate the model's performance 

```python
y_pred = modelv1.predict(x_test)
print(y_pred) 
```

### This displays the actual target values from the test dataset, which are used to compare with the predicted values.

```python
print(y_test)
```

### In this step, the performance of the Linear Regression model is evaluated using different metrics. Mean Absolute Error (MAE) measures the average error, Mean Squared Error (MSE) penalizes larger errors, and the R² Score indicates how well the model fits the data (values closer to 1 represent better performance).

## EVALUATION METRICS OF LINEAR REGRESSION : 

```python
print("EVALUATION METRICS OF LINEAR REGRESSION : ")
print("Mean Absolute Error = ",mean_absolute_error(y_test,y_pred))
print("Mean Square Error = ",mean_squared_error(y_test,y_pred))
print("R2 Score =",r2_score(y_test,y_pred))
```

# Decision Tree Algorithm 

### Import Libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb 
import sklearn 

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

### Load Dataset

```python
data2 = pd.read_csv("student_data.csv")
```

### Info of the Dataset

```python
data2.info()
```

### Y1 contains the Targets of the Dataset 

```python
y1 = data2["G3"]          # target
```

### X1 contains the Features of the Dataset

```python
x1 = data2.drop("G3", axis=1)  #features
```

### Categorical features are converted into numerical format using one-hot encoding

```python
x1 = pd.get_dummies(x1, dtype=int) 
# This ensures that all input features are in a suitable form for machine learning models, which require numerical data. So,that model can understand.
```

### The dataset is split into training (80%) and testing (20%) sets. The training set is used to train the model, while the testing set is used to evaluate its performance with Predicted data of the Model.

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.2,random_state=0)
```

### The Decision Tree Classifier model is trained using the training data to learn the relationship between input features and the target variable.

```python
dt_classifier = DecisionTreeClassifier(random_state = 42)  #Decision tree classifier
dt_classifier.fit(x1_train,y1_train)  #Train the model
```

### Now, y_pred have the Predicted values of the model . So, now these predicted values are used to evaluate the model's performance.

```python
y1_pred = dt_classifier.predict(x1_test) #make prediction using DT classifier
print(y1_pred)
```

### This displays the actual target values from the test dataset, which are used to compare with the predicted values.

```python
print(y1_test)
```

### In this step, the performance of the Decision Tree model is evaluated using standard regression metrics. Mean Absolute Error (MAE) represents the average prediction error, Mean Squared Error (MSE) gives higher importance to larger errors, and the R² Score indicates how well the model explains the variability in the data.

## EVALUATION METRICS OF DECISION TREE :

```python
print("EVALUATION METRICS OF DECISION TREE : ")
print("Mean Absolute Error = ",mean_absolute_error(y1_test,y1_pred))
print("Mean Square Error = ",mean_squared_error(y1_test,y1_pred))
print("R2 Score =",r2_score(y1_test,y1_pred))
```

# Random Forest Algorithm 

### Import Libraries 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn 

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

### Load Dataset

```python
data3 = pd.read_csv("student_data.csv")
```

### Info of the Dataset

```python
data3.info()
```

### Y2 Contains the Targets of the dataset

```python
y2 = data3["G3"]    #Targets
```

### X2 Contains the Features of the Dataset

```python
x2 = data3.drop("G3", axis=1)  #features
```

### Categorical features are converted into numerical format using one-hot encoding

## EVALUATION METRICS OF DECISION TREE :
